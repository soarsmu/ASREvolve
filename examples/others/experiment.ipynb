```python
res_template = "output/europarl-seed2021/result/rv/deepspeech_deepspeech2_wav2letter_wit_wav2vec2/num_iteration_5/text_batch_size_{}/with-estimator-facebook-bart-base.json"

def get_nftc(fpath) :
    with open(fpath) as f:
        data = json.load(f)
    return data["number_of_failed_test_cases_all"][-1]

visibilities = []
res = []
# for text_batch_size in [400, 600,  800, 1000, 1200, 1400, 1600, 1800, 2000, 2400, 3000, 5000, 10000, 15000] :
for text_batch_size in range(400, 8001, 400) :
    fpath = res_template.format(text_batch_size)
    res.append(get_nftc(fpath))
    visibilities.append(text_batch_size)
    
```

```python
result_folder = "output/europarl-seed2021/result/rv/deepspeech_deepspeech2_wav2letter_wit_wav2vec2/num_iteration_5/text_batch_size_1200/"
estimators = ["albert-base-v2", "facebook/bart-base", "bert-base-cased", "bert-base-uncased", "distilbert-base-uncased", "xlnet-base-cased", "roberta-base"]
estimators = ["albert-base-v2", "facebook/bart-base", "bert-base-uncased", "distilbert-base-uncased", "xlnet-base-cased", "roberta-base"]

# estimators = ["albert-base-v2", "facebook/bart-base", "bert-base-cased", "bert-base-uncased", "distilbert-base-uncased"]


def rename(s):
    return s.replace("/", "-")

estimators = [rename(e) for e in estimators]
```

```python
def get_nftc(fpath) :
    with open(fpath) as f:
        data = json.load(f)
    return data["number_of_failed_test_cases_all"][-1]

estimator = "albert-base-v2"
fpath = result_folder + "with-estimator-" + estimator + ".json"
get_nftc(fpath)
```

```python
nftcs = []

for estimator in estimators :
    fpath = result_folder + "with-estimator-" + estimator + ".json"
    nftcs.append(get_nftc(fpath))

fpath = result_folder + "without-estimator.json"
nftcs.append(get_nftc(fpath))

final_estimators = estimators + ["No estimator"]

df = pd.DataFrame(data={"estimator": final_estimators, "failed test cases": nftcs})
df = df.sort_values(by=["failed test cases"]).reset_index(drop=True)
df
```

```python
def print_rq2_tex(estimators, nftcs) :
    for e, n in zip(estimators, nftcs) :
        print(f"{e} & {n} \\\\")

print_rq2_tex(df["estimator"], df["failed test cases"])
```
